```python
## Assignments
#divide my identification number with 6 to find out my task

def find_remainder(id_number):
  
  try:
    id_number = int(id_number)  
    remainder = id_number % 6
    return remainder
  except ValueError:
    return "Invalid input. Please enter an integer."


id_number_str = input("Enter my identification number: ")

remainder = find_remainder(id_number_str)
print(f"The remainder when my ID is divided by 6 is: {remainder}")
print(f"So i am assigned to complete the task:{remainder}th")
```

```python
#task-4:Create 'Number of relatives' column and Filter for passengers who survived

import pandas as pd
import matplotlib.pyplot as plt


try:
  titanic_df = pd.read_csv('titanic.csv')
except FileNotFoundError:
  print("Error: titanic.csv not found. Please upload the file to the current directory.")
  titanic_df = None

if titanic_df is not None:
  
  titanic_df['Number of relatives'] = titanic_df['SibSp'] + titanic_df['Parch']
  
  
  survived_passengers = titanic_df[titanic_df['Survived'] == 1]
  
  
  plt.figure(figsize=(10, 6))
  plt.scatter(survived_passengers['Number of relatives'], survived_passengers['Fare'], alpha=0.5)
  plt.title('Fare vs. Number of Relatives for Survived Passengers')
  plt.xlabel('Number of Relatives')
  plt.ylabel('Fare')
  plt.grid(True)
  plt.show()
  
```

```python
#task-6:
import pandas as pd
import matplotlib.pyplot as plt


try:
  df = pd.read_csv('titanic.csv')
except FileNotFoundError:
  print("Error: titanic.csv not found. Please upload the file to your Colab environment.")
 
else:
  
  df['Number of relatives'] = df['SibSp'] + df['Parch']

  
  survived_passengers = df[df['Survived'] == 1]

  
  plt.figure(figsize=(10, 6))
  plt.scatter(survived_passengers['Number of relatives'], survived_passengers['Age'],alpha=0.7)
  plt.xlabel('Number of Relatives')
  plt.ylabel('Age')
  plt.title('Scatter Plot of Survived Passengers: Age vs. Number of Relatives')
  plt.grid(True)
  plt.show()
```
